# 01. Análise Exploratória de Dados (EDA)

Este notebook é dedicado à exploração inicial do **Student Depression Dataset**. O objetivo é compreender a distribuição da variável-alvo (`Depression`), identificar o volume de dados ausentes e analisar o comportamento das variáveis numéricas e categóricas, incluindo as features geradas pelo módulo de engenharia de atributos.

```python
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path

# Permite importar os módulos da pasta src/ estando dentro da pasta notebooks/
root_path = Path.cwd().parent
if str(root_path) not in sys.path:
    sys.path.append(str(root_path))

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Importando os módulos do pipeline do projeto
from src.data.make_dataset import load_raw_data
from src.features.build_features import engineer_features

# Configuração visual dos gráficos
sns.set_theme(style="whitegrid", context="notebook")
```

```python
# Caminhos relativos à raiz do projeto
CSV_FILE = root_path / "data" / "raw" / "Student Depression Dataset.csv"

# Carrega os dados brutos usando o módulo correto
df_raw = load_raw_data(CSV_FILE)

# Aplica as transformações linha a linha do módulo de features
df = engineer_features(df_raw)

print(f"Dataset bruto: {df_raw.shape}")
print(f"Dataset com features derivadas: {df.shape}")
df.head()
```

## 2. Análise de Dados Ausentes

Verificação de valores nulos mapeados nas colunas originais do dataset.

```python
missing = df_raw.isna().sum().sort_values(ascending=False)
missing = missing[missing > 0]

plt.figure(figsize=(8, 3.6))
if len(missing) > 0:
    ax = sns.barplot(x=missing.values, y=missing.index, color="#0f766e")
    ax.set_title("Dados Ausentes por Atributo Original")
    ax.set_xlabel("Quantidade")
    ax.set_ylabel("")
    for i, v in enumerate(missing.values):
        ax.text(v + 0.05, i, str(v), va="center")
else:
    plt.text(
        0.5,
        0.5,
        "Não foram encontrados valores ausentes no dataset bruto.",
        ha="center",
        va="center",
    )
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## 3. Distribuição das Variáveis Numéricas e Atributos Derivados

Análise do comportamento das principais variáveis contínuas e ordinais divididas por classe.

```python
num_plot_cols = [
    "Age",
    "Academic Pressure",
    "Financial Stress",
    "Work/Study Hours",
    "Pressure Satisfaction Ratio",
    "Stress Academic Index",
]

fig, axes = plt.subplots(2, 3, figsize=(14, 7))
axes = axes.ravel()

for ax, col in zip(axes, num_plot_cols):
    if col in df.columns:
        sns.histplot(
            data=df,
            x=col,
            hue="Depression",
            bins=24,
            multiple="stack",
            palette=["#2563eb", "#dc2626"],
            ax=ax,
        )
        ax.set_title(col)
        ax.set_xlabel("")

fig.suptitle("Distribuições Numéricas e Features Derivadas", y=1.02)
plt.tight_layout()
plt.show()
```

## 4. Matriz de Correlação Linear

Avaliando a correlação de Pearson entre as variáveis numéricas originais, os scores derivados e a variável-alvo.

```python
numeric_features = df.select_dtypes(exclude="object").columns.tolist()
corr = df[numeric_features].corr(numeric_only=True)

plt.figure(figsize=(12, 9))
sns.heatmap(corr, cmap="RdBu_r", center=0, annot=False, cbar_kws={"shrink": 0.75})
plt.title("Matriz de Correlação Linear com Features Derivadas")
plt.tight_layout()
plt.show()
```

## 5. Proporção da Classe Positiva por Categoria

Análise do impacto de variáveis categóricas (como histórico familiar e pensamentos suicidas) na taxa real de depressão.

```python
cat_rate_cols = [
    "Gender",
    "Sleep Duration",
    "Dietary Habits",
    "Have you ever had suicidal thoughts ?",
    "Family History of Mental Illness",
]

fig, axes = plt.subplots(
    len(cat_rate_cols), 1, figsize=(11, 3.1 * len(cat_rate_cols))
)

for ax, col in zip(axes, cat_rate_cols):
    if col in df.columns:
        rates = df.groupby(col)["Depression"].mean().sort_values(ascending=False)
        sns.barplot(x=rates.values, y=rates.index, palette="viridis", ax=ax)
        ax.set_xlim(0, 1)
        ax.set_title(f"Taxa de Depression=1 por {col}")
        ax.set_xlabel("Proporção Positiva")
        ax.set_ylabel("")
        for i, v in enumerate(rates.values):
            ax.text(min(v + 0.015, 0.98), i, f"{v:.1%}", va="center")

plt.tight_layout()
plt.show()
```
